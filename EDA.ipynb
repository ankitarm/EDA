```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame

# Extract the quarter from the 'ORDERDATE' column
df_2['Quarter'] = df_2['ORDERDATE'].dt.quarter

# Group by 'Quarter' and calculate the total sales (sum of 'price') for each quarter
quarterly_sales = df_2.groupby('Quarter')['price'].sum()

# Plotting the total sales for each quarter
plt.figure(figsize=(8, 6))
quarterly_sales.plot(kind='bar')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.title('Total Sales by Quarter')
plt.xticks(rotation=0)  # Rotate x-axis labels if necessary
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame

# Group by 'category' and 'Month' and calculate the total sales (sum of 'price') for each category and month
category_monthly_sales = df_2.groupby(['category', 'Month'])['price'].sum().unstack()

# Plotting the trend of sales for each category
plt.figure(figsize=(12, 8))

for category in category_monthly_sales.index:
    plt.plot(category_monthly_sales.columns, category_monthly_sales.loc[category], label=category)

plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Trend of Sales by Category')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(range(1, 13))  # Assuming months are represented by numbers 1-12
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract the year from the 'ORDERDATE' column
df_2['Year'] = df_2['ORDERDATE'].dt.year

# Group by 'category' and 'Year' and calculate the total sales (sum of 'price') for each category and year
category_yearly_sales = df_2.groupby(['category', 'Year'])['price'].sum()

# Plotting the sales trend for each category over the years
plt.figure(figsize=(12, 8))

# Iterate over unique categories
for category in df_2['category'].unique():
    # Extract sales data for the current category
    category_sales = category_yearly_sales[category]
    
    # Plot sales trend for the current category
    plt.plot(category_sales.index.get_level_values('Year'), category_sales.values, label=category)

plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Sales Trend by Category Over the Years')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
